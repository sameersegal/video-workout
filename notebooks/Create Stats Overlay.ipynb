```python
time = []
heartratebpm = []
cadence = []
power = []

for child in root.findall('.//garmin_v2:Trackpoint',ns):
    t = child.find("garmin_v2:Time",ns).text
    h = child.find("garmin_v2:HeartRateBpm",ns)[0].text
    c = child.find("garmin_v2:Cadence",ns).text
    p = child.find("garmin_v2:Extensions",ns)[0][1].text
    time.append(t)
    heartratebpm.append(h)
    cadence.append(c)
    power.append(p)
    
df = pd.DataFrame({'time':time,'heartratebpm':heartratebpm,'cadence':cadence,'power':power})
df['time'] = pd.to_datetime(df['time'])
df['time'] = df['time'] - df['time'][0]
df['time'] = df['time'].astype(str).str.split().str[2]
df['heartratebpm'] = pd.to_numeric(df['heartratebpm'])
df['cadence'] = pd.to_numeric(df['cadence'])
df['power'] = pd.to_numeric(df['power'])
df.loc[df['power'] == 0,'power'] = np.NAN
df['time'].drop_duplicates(inplace=True)
df.set_index('time')
```

```python
df.plot(x='time',y=['heartratebpm','cadence','power'])
```
